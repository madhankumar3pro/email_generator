Cleaning text data

````python
def clean_text(text):
    """
    Cleans the given text by removing unwanted characters, words, and stop words,
    and applies lemmatization.
    """

    nltk.download('stopwords', force=True)
    nltk.download('wordnet', force=True)
    # Remove non-alphanumeric characters
    text = re.sub(r'[^a-zA-Z0-9.\s]', '', text)
    # Load stop words
    STOP_WORDS = set(stopwords.words('english'))
    # Split into words, remove stop words, and lemmatize
    cleaned_words = [
        word.lower() # convert to lowercase
        for word in text.split()
        if word.lower() not in STOP_WORDS  # Remove stop words
    ]
    
    return ' '.join(cleaned_words[0:400])
````

Exctracting text data in url's

````python
def extract_headings_paragraphs(url):
  try:
    heading_info = []
    # Send an HTTP GET request to the URL
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception for HTTP errors

    # Parse the page content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract all headings (h1, h2, h3, h4, h5, h6)
    headings = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6','p'])
    for heading in headings:
        heading_info.append(heading.get_text(strip=True))
    
    return heading_info

  except requests.exceptions.RequestException as e:
    return f"{e}"
````
